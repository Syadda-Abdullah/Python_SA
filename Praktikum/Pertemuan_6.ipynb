**Menghitung Gaji Rata Rata Karyawan**

```python
#Fungsi Menghitung gaji
def GajiTengah(salary):
    min_salary = min(salary)
    max_salary = max(salary)
    total_salary = sum(salary) - min_salary - max_salary
    # Mengurangkan 2 agar gaji minimum dan maksimum tidak diambil
    average = total_salary / (len(salary) - 2)
    return average

# Input gaji list karyawan
input_salaries = input("Masukkan gaji karyawan (pisahkan dengan spasi): ")
salaries = list(map(int, input_salaries.split()))

# Hitung rata-rata gaji
result = GajiTengah(salaries)

print(f"Gaji rata-rata tidak termasuk gaji minimum dan maksimum: {result:.2f}")
```

**Menentukan Angka Keberuntungan**

```python
def luckyNumbers(matrix):
    # Menemukan minimum dalam setiap baris
    min_in_row = [min(row) for row in matrix]  # Menyimpan nilai minimum dalam setiap baris

    # Menemukan maksimum dalam setiap kolom
    max_in_col = [max(col) for col in zip(*matrix)]  # Menyimpan nilai maksimum dalam setiap kolom

    # Menemukan angka keberuntungan dengan membandingkan minimum baris dan maksimum kolom
    lucky_numbers = [num for num in min_in_row if num in max_in_col]  # Mengidentifikasi angka keberuntungan

    return lucky_numbers

# Input matriks sebagai list 2D
input_matrix = input("Masukkan matriks (Contoh [[a,b,c],[p,q,r]]): ")
matrix = eval(input_matrix)  # Menguraikan matriks dari input pengguna

# Hitung angka keberuntungan dalam matriks
result = luckyNumbers(matrix)

# Menampilkan angka-angka keberuntungan
print("Angka-angka keberuntungan:", result)
```

**Membuat Graf Saham**

```python
import matplotlib.pyplot as plt

def maxProfit(prices):
    max_profit = 0
    buy_price = float('inf')
    sell_price = 0
    buy_day = 0
    sell_day = 0

    for i in range(len(prices)):
        if prices[i] < buy_price:
            buy_price = prices[i]
            buy_day = i

        if prices[i] - buy_price > max_profit:
            max_profit = prices[i] - buy_price
            sell_price = prices[i]
            sell_day = i

    return max_profit, buy_day, sell_day

# Input harga saham sebagai list
input_prices = input("Masukkan harga saham (pisahkan dengan spasi): ")
prices = list(map(int, input_prices.split()))

# Hitung keuntungan maksimum
max_profit, buy_day, sell_day = maxProfit(prices)

if max_profit > 0:
    print(f"Keuntungan maksimum yang bisa didapat adalah: {max_profit}")
    print(f"Membeli pada hari {buy_day + 1} dengan harga {prices[buy_day]} dan menjual pada hari {sell_day + 1} dengan harga {prices[sell_day]}")

    # Plot harga saham
    plt.plot(prices, label='Harga Saham', marker='o')
    plt.plot(buy_day, prices[buy_day], 'go', label='Beli')
    plt.plot(sell_day, prices[sell_day], 'ro', label='Jual')
    plt.xlabel('Hari')
    plt.ylabel('Harga Saham')
    plt.title('Grafik Harga Saham')
    plt.legend()
    plt.show()
else:
    print("Tidak ada keuntungan yang bisa didapat.")
```
