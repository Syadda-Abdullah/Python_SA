**Membaca Data CSV**

```python
import pandas as pd
import matplotlib.pyplot as plt

#baca datanya
file_path = 'TugasP7.csv'
data = pd.read_csv(file_path)

#print data mentahnya
print("Data Mentah:")
print(data.to_string())

#konversi tipe data kolom 'produksi_padi' ke numerik
data['produksi_padi'] =pd.to_numeric(data['produksi_padi'].str.replace(',', ''),errors='coerce')

#bersihin data yang kosong
data.dropna(inplace=True)

#print data setelah diolah
print("\nInfo DataFrame setelah pengolahan:")
print(data)
```

**Menentukan Data Tertinggi Dan Terendah**

```python
#itung total produksi padi di setiap kelurahan
total_produksi_kelurahan = data.groupby('kode_kelurahan')['produksi_padi'].sum()

#identifikasi 5 kelurahan dengan total produksi padi tertinggi dan terendah
top5 = total_produksi_kelurahan.nlargest(5)
bottom5 = total_produksi_kelurahan.nsmallest(5)

print("\n5 Kelurahan dengan Produksi Padi Tertinggi:")
print(top5)

print("\n5 Kelurahan dengan Produksi Padi Terendah:")
print(bottom5)
```

**Membandingkan Data Dengan Visualisasi Graph 1**

```python
# Langkah 3: Visualisasi Data
# Buat grafik batang untuk 5 kelurahan dengan total produksi padi tertinggi
plt.figure(figsize=(10, 5))
plt.bar(top5.index, top5.values, color='green')
plt.title('5 Kelurahan dengan Total Produksi Padi Tertinggi')
plt.xlabel('Kelurahan')
plt.ylabel('Total Produksi Padi')
plt.xticks(rotation=45)
plt.show()

# Buat grafik batang untuk 5 kelurahan dengan total produksi padi terendah
plt.figure(figsize=(10, 5))
plt.bar(bottom5.index, bottom5.values, color='red')
plt.title('5 Kelurahan dengan Total Produksi Padi Terendah')
plt.xlabel('Kelurahan')
plt.ylabel('Total Produksi Padi')
plt.xticks(rotation=45)
plt.show()
```

**Membandingkan Data Dengan Visualisasi Grap 2**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
file_tugas = 'Data/TugasP7.csv'
data = pd.read_csv(file_tugas)

# Pengecekan dan pembersihan data
data['produksi_padi'] = pd.to_numeric(data['produksi_padi'], errors='coerce')
data_cleaned = data.dropna()

# Analisis statistik
total_produksi_per_kelurahan = data_cleaned.groupby('nama_kelurahan')['produksi_padi'].sum()

# Mengambil 5 kelurahan dengan total produksi padi tertinggi
top5_kelurahan = total_produksi_per_kelurahan.sort_values(ascending=False).head(5)

# Mengambil 5 kelurahan dengan total produksi padi terendah
bottom5_kelurahan = total_produksi_per_kelurahan.sort_values().head(5)

# Plotting grafik batang untuk 5 kelurahan tertinggi
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
top5_kelurahan.plot(kind='bar', color='skyblue')
plt.title('Top 5 Kelurahan - Total Produksi Padi')
plt.xlabel('Nama Kelurahan')
plt.ylabel('Total Produksi Padi')

# Plotting grafik batang untuk 5 kelurahan terendah
plt.subplot(1, 2, 2)
bottom5_kelurahan.plot(kind='bar', color='salmon')
plt.title('Bottom 5 Kelurahan - Total Produksi Padi')
plt.xlabel('Nama Kelurahan')
plt.ylabel('Total Produksi Padi')

plt.tight_layout()
plt.show()
```
